```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By

class LiveCareer:
    def __init__(self, keyword:str):
        self.keyword = keyword
        self.driver = webdriver.Chrome()
        url = 'https://www.livecareer.com/resume-search/search?jt=' + keyword.replace(' ', '+')
        self.driver.get(url)
        time.sleep(1)
        self.page_num = int(self.driver.find_elements(By.CLASS_NAME, 'sc-1uorqye-1')[-1].text)
        self.searched_urls = []
        if self.page_num > 1:
            for i in range(self.page_num):
                self.searched_urls.append(url + '&page=' + str(i+1))
        self.driver.quit()
        self.resumes = pd.DataFrame(columns=['Index',
                                             'Resume', 
                                             'Job_Title', 
                                             'Company_Location', 
                                             'Position', 
                                             'Last_Updated', 
                                             'Resume_URL'],
                                             index=['Index']
                                             )
        self.get_resume_urls()

    def get_resume_urls(self):
        if len(self.searched_urls) == 0:
            print("Searched pages = 0!")
            pass
        else:
            for id, url in enumerate(self.searched_urls):
                print("Current page: ", url)
                self.driver = webdriver.Chrome()
                self.driver.get(url)
                time.sleep(1)
                resumes = self.driver.find_elements(By.CLASS_NAME, 'sc-1dzblrg-0')
                self.get_resume_info(resumes, id)
                self.driver.quit()
                # Print out the driver status
                print("Quited the driver!")

    def get_resume_info(self, resumes, id):
        for i, resume in enumerate(resumes):
            # Extract the resume URL (href)
            resume_url = resume.get_attribute('href')
            # Extract job title
            job_title = resume.find_element(By.CLASS_NAME, 'title').text
            # Extract company and location
            company_location = resume.find_element(By.CLASS_NAME, 'company').text
            # Extract the position and company title
            company_title = resume.find_element(By.CLASS_NAME, 'company-title').text
            # Extract the last updated date
            last_updated = resume.find_element(By.CLASS_NAME, 'last-updated').text
            # Store the extracted information to dataframe 
            _df = pd.DataFrame({'Index': f"{self.keyword.replace(' ','')}-{id+1}-{i+1}", 
                                'Resume': i+1, 
                                'Job_Title': job_title, 
                                'Company_Location': company_location, 
                                'Position': company_title, 
                                'Last_Updated': last_updated, 
                                'Resume_URL': resume_url}, 
                                index=['Index']
                                )
            # Concatenate the extracted information to the main dataframe
            self.resumes = pd.concat([self.resumes, _df])

    def save_to_csv(self, location:str):
        self.resumes.drop_duplicates().dropna().to_csv(location, index=False)
# Execute the scraper
# LC = LiveCareer(keyword='data engineer')

# Save the extracted information to csv
# LC.save_to_csv(location='dataengineer_resume_urls.csv')
```
