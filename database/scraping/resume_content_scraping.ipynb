```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By

class ResumeScraper:
    def __init__(self, url):
        self.url = url
        self.driver = webdriver.Chrome()
        self.driver.get(self.url)
        time.sleep(1)

        self.resume_html = self.driver.find_elements(By.CLASS_NAME, 'document')
        self.headings = self.driver.find_elements(By.CLASS_NAME, 'heading') 
        self.singlecolumns = self.driver.find_elements(By.CLASS_NAME, 'singlecolumn') 

        self.resume_text = []
        for html in self.resume_html:
            self.resume_text.append(html.text)
        self.resume_text = '------------------------'.join(self.resume_text)
        print(self.resume_text)

        self.page_source = self.get_all_html()
        self.driver.quit()
    
    def get_all_html(self):
        return self.driver.page_source

url = 'https://www.livecareer.com/resume-search/r/credentialing-provider-data-analyst-4362c96c54f04a578ed3d22d6d8563a3'
RC = ResumeScraper(url)
```
